#  Condicionales (Decisiones)

Los condicionales en Python nos permiten tomar decisiones en un programa. Con ellos, el código puede ejecutar diferentes instrucciones dependiendo de si ciertas condiciones se cumplen o no.
Los condicionales funcionan evaluando expresiones lógicas, que pueden ser verdaderas (True) o falsas (False). Dependiendo del resultado, se ejecuta una parte del código u otra.

existen 3 tipos de condicionales:
- *if*: Se usa para ejecutar código solo si se cumple una condición.
- *elif*: Se usa cuando hay varias condiciones posibles.
- *else*: Se ejecuta cuando ninguna de las condiciones anteriores se cumplen 

***ejemplo:***
El Portero del Club
- if (Si...): Es el portero que revisa tu edad.
- elif (Si no, pero si...): Si no tienes la edad, revisa si tienes invitación VIP.
- else (De lo contrario): Si no cumples nada, te quedas fuera.

**importante**: 
- La Tabulación/Sangría después de la instrucción if indica que cosas se realizarán si se cumple la condición. Esto se llama Indentación. le dice a Python qué instrucciones están "dentro" del condicional.
- Los dos puntos (:) Siempre van al final de la condición.

```python
# sintaxis del if

edad = int(input("Ingresa tu edad: "))
if edad >= 18:
    print("Eres mayor de edad")  # Se ejecuta si edad es 18 o más
```

```python
# siintaxis if-else

temperatura = int(input("Ingresa la temperatura actual: "))
print(" Temperatura ingresada:", temperatura)
if temperatura > 20:
    print("Hace calor")  # Se ejecuta si temperatura es mayor a 20
else:
    print("Hace frío")  # Se ejecuta si temperatura es 20 o menor
```

```python
#sintaxis if-elif-else

nota = int(input("Ingresa tu nota: "))
print(" Nota ingresada:", nota)
if nota >= 90:
    print("Excelente")
elif nota >= 70:
    print("Bueno")
elif nota >= 55:
    print("Regular")
else:
    print("Malo")
```

## Operadores 

En las condicines se pueden utilizar operadores logicos y Operadores relacionales, o de comparación, que permitir construir expreciones que reepresentan condiciones complejas.

#### Operadores relacionales, o de comparación
Comparan expresiones de cualquier tipo. El resultado siempre es un valor booleano: True o False.
| Operador | Significado         | Ejemplo             |
|----------|---------------------|---------------------|
|==        | Igual que           |5 == 5 → True        |
|!=        | Distinto a          |4 != 2 → True        |
|<         | Menor que           |3 < 8 → True         |
|>         | Mayor que           |10 > 5 → True        |
|<=        |<=	Menor o igual que|6 <= 6 → True        |
|>=        |Mayor o igual que    |7 >= 8 → False       |


#### Operadores lógicos
Permiten construir expresiones que representan condiciones complejas a partir de otras expresiones booleanas.
- Conjunción lógica and - False si cualquiera de sus operandos es False, True si todos son True.
- Disyunción lógica or - True si cualquiera de sus operandos es True, False si todos son False.
- Negación lógica not - True cuando el operando es False, y viceversa.

| Operador | Significado         | Ejemplo             |
|----------|---------------------|---------------------|
| and      | Y lógico            | x > 0 and x < 10    |
| or       | O lógico            | x == 0 or y == 0    |
| not      | Negación lógica     | not x == y          |


```python
edad = 20
tiene_entrada = True
if edad >= 18 and tiene_entrada:
    print("Puedes entrar")

#----------------------------------------------
print("-"*20)

numero = 7
if numero % 2 == 0:
    print("Es par")
else:
    print("Es impar")
```

### Resumen rápido

- if → una condición
- if - else → dos caminos
- if - elif - else → múltiples opciones
- Usa operadores (==, >, etc.) para comparar
- Usa and, or, not para combinar
