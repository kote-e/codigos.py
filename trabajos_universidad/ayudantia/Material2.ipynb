# Llamados a funciones y uso de biblioteca

En Python hay dos tipos de funciones que puedes usar:
- Incorporadas: Ya vienen "en tu mochila" (ej. abs(), len(), round()).
- De biblioteca: Están en una "bodega" y debes buscarlas usando import (ej. math, random).

#### *Funciones que ya vienen listas en Python*
Son funciones que puedes usar directamente, sin hacer nada antes.

Ejemplos:
- round() → redondea números
- abs() → devuelve el valor absoluto (sin signo negativo)
- len() → devuelve el largo de una cadena
No necesitas preparar nada, Python ya las tiene incluidas.

#### *Funciones que están en bibliotecas*

Hay otras funciones que no vienen activadas por defecto, sino que están dentro de bibliotecas (o módulos).
Una biblioteca es como una caja de herramientas con muchas funciones relacionadas.

Por ejemplo:
- math → funciones matemáticas
- random → funciones para números aleatorios
Para usarlas primero debes importarlas.

##### *Diferencia:*

- utilizamos ``` import math ``` cuando 

````python
#Funciones incorporadas en Python

print(round(3.7))   # resultado: 4, redondea al número entero más cercano
print(abs(-5))      # resultado: 5, devuelve el valor absoluto de un número
print(len("Hola"))  # resultado: 4, devuelve la longitud de la cadena de texto

print(40*"-")  # imprime una línea de 40 guiones para separar secciones

#Funciones de biliotecas

import math # importa la biblioteca matemática para usar sus funciones matemáticas avanzadas

print(math.sqrt(16))  # raíz cuadrada → 4, primero se importa la biblioteca math y luego se llama a la función sqrt para calcular la raíz cuadrada
print(math.pi)  # valor de pi → 3.141592653589793, accede a la constante pi definida en la biblioteca math

print(40*"-")

from math import exp  # importa solo la función exp de la biblioteca math para calcular la función exponencial
Exp= exp(1)  #aqui no se importa toda la biblioteca math, sino solo la función exp, lo que permite usarla directamente sin el prefijo math. 
print(Exp)  # calcula el valor de e(euler) elevado a la potencia de 1, donde e es la base del logaritmo natural, aproximadamente igual a 2.71828

print(40*"-")

import random # importa la biblioteca random para usar sus funciones de generación de números aleatorios
print(random.randint(1, 10))  # número aleatorio entre 1 y 10, genera un número entero aleatorio entre 1 y 10 (inclusive)
````

## Precedencia de Operadores

Cuando Python resuelve una expresión matemática, **no calcula todo de izquierda a derecha**.  
Sigue un **orden de prioridad**, llamado **precedencia de operadores**.

Los operadores con **mayor prioridad se calculan primero**.

### Orden de precedencia (de mayor a menor)

| Prioridad | Operador | Qué hace |
|-----------|----------|----------|
| 1 | `( )` | Paréntesis |
| 2 | `abs()`, `sqrt()`, `randint()` | Llamadas a funciones, aun    ue estas no sean operadores |
| 3 | `**` | Potencia |
| 4 | `+x`, `-x` | Signo positivo o negativo |
| 5 | `*`, `/`, `//`, `%` | Multiplicación, división, división entera, módulo |
| 6 | `+`, `-` | Suma y resta |

💡 **Regla simple:**  
Python sigue casi el mismo orden que en matemáticas:
1. Paréntesis  
2. Potencias  
3. Multiplicación, división y Módulo
4. Suma y resta  

**Tips:**  
- Si no estás segura del orden, **usa paréntesis**, porque Python siempre los calcula primero.
- **El operador Módulo (%):** A menudo se confunde con el porcentaje, pero en programación es el resto de la división. (ejemplo: si repartes 10 galletas entre 3 niños, a cada uno le das 3 y te sobra 1. Ese 1 es el módulo.)

````python
# 1. Paréntesis ()
resultado_parentesis = (2 + 3) * 4
print("Paréntesis:", resultado_parentesis)  # 20

print(40*"-")

# 2. Llamadas a funciones
import math
import random

print("abs(-5):", abs(-5))              # 5
print("sqrt(16):", math.sqrt(16))      # 4.0
print("randint(1,10):", random.randint(1, 10))  # número aleatorio entre 1 y 10

print(40*"-")

# 3. Potencia **
resultado_potencia = 2 ** 3
print("Potencia:", resultado_potencia)  # 8

print(40*"-")

# 4. Operador unario (+x, -x)
numero = 5
print("Unario positivo:", +numero)  # 5
print("Unario negativo:", -numero)  # -5

print(40*"-")

# 5. Multiplicación, división, división entera y módulo
print("Multiplicación:", 6 * 3)   # 18
print("División:", 10 / 2)        # 5.0
print("División entera:", 10 // 3) # 3
print("Módulo (resto):", 10 % 3)   # 1

print(40*"-")

# 6. Suma y resta
print("Suma:", 5 + 2)   # 7
print("Resta:", 5 - 2)  # 3
````

## Conversión entre tipos de datos (Casting)

El **casting** es cuando cambiamos un dato de **un tipo a otro**.

Por ejemplo, convertir:
- un número decimal a entero
- un texto a número
- un número a texto
En Python esto se hace usando funciones especiales.

un ejemplo donde usamos esto es en la función input() siempre recibe texto (str). Si quieres hacer cálculos, debes "traducirlo":
- int(input()): Traduce el texto a número entero.
- float(input()): Traduce el texto a número decimal.

*Ejemplos*

````python
# Conversión entre tipos de datos (Casting)

# Decimal -> Entero
numero_entero = int(3.5)
print(numero_entero)   # Resultado: 3

# Texto -> Decimal
numero_decimal = float("1")
print(numero_decimal)  # Resultado: 1.0

# Número -> Texto
texto = str(25)
print(texto)           # Resultado: '25'

# Número -> Booleano
valor_booleano = bool(0.0)
print(valor_booleano)  # Resultado: False
````


### Tipos de datos comunes

| Tipo | Qué representa | Ejemplo |
|-----|---------------|--------|
| `int` | Números enteros | `5`, `10`, `-3` |
| `float` | Números con decimales | `3.5`, `0.2` |
| `str` | Texto | `"hola"` |
| `bool` | Verdadero o falso | `True`, `False` |


Regla fácil de recordar:
- int() → convierte a entero
- float() → convierte a decimal
- str() → convierte a texto
- bool() → convierte a verdadero o falso

## Entrada de datos (input())

En Python, la función input() se usa para pedirle información al usuario mientras el programa se está ejecutando.
Lo que la persona escriba se guarda en una variable.

````python
dato = input()
print("El dato ingresado es:", dato)
````

También podemos mostrar un mensaje para que el usuario sepa qué debe escribir.

````python
nombre = input("Ingrese su nombre: ")
print("Hola, " + nombre + "! Bienvenido a la clase de Python.")
````

*¿Qué tipo de dato es nombre?*

El tipo de dato es texto, es decir un string (str).
Esto pasa porque input() siempre devuelve texto, aunque el usuario escriba números.

*Ejemplo para comprobarlo:*

````python
nombre = input("Ingrese su nombre: ")

print(nombre)
print(type(nombre))
````

*Algo importante para recordar:* Si el usuario escribe un número, igual será texto.

Si quieres usarlo como número, debes convertirlo (casting)

````python
edad = input("Ingrese su edad: ")
print("Su edad es:", edad)
print(type(edad))

print(40*"-")

edad = int(input("Ingrese su edad: "))
print("Su edad es:", edad)
print(type(edad))
````
