## Variables 

Las variables son espacios en la memoria que almacenan datos, al que se le asigna un valor o texto.
(en Python no es necesitas declarar el tipo de variable se asigna el tipo automáticamente según lo que guardemos.)

**Tipos de datos básicos:**

| Tipo            | Ejemplo      | Que es?                                                                      |
| --------------  | ------------ | -----------------------------------------------------------------------------|
| Entero (int)    | 5            | Son basicamente numeros                                                      |  
| Flotante (float)| 3.14         | Permite usar numeros decimales                                               |
| Cadena (str)    | "Hola"       | Nos permite crear cadenas de texto                                           |
| Booleano (bool) | True / False | Es un tipo de dato que trabaja con dos estados, True y False                 |

*Ejemplo completo:* 

```python
# Esto es un comentario

#Tipo de dato int (entero)
print("int")
x = 1
y = 2
z = 200
print(x + y + z)

print (40*"-")  #imprime una linea de guiones para separar secciones

#Tipo de dato float (decimal)

print("float")
f1 = 3.14
print(f1)
foo = x + f1
print(type(foo)) # el tipo de dato de foo es float porque al sumar un int con un float el resultado es un float

print (40*"-")  

#Tipo de dato str (cadena de texto)
print("str")
s1 = 'Hola mundo!' # las comillas simples o dobles sirven para definir un string
print(s1)

print(s1[5:]) #imprime desde el indice 5 hasta el final del string
print(s1[6:8]) #imprime desde el indice 6 hasta el indice 8 (sin incluir el indice 8)
s2 = s1 + '!!' # concatenacion de strings, se puede usar el operador + para unir dos strings
print(s2)       
print(f'El largo del string es: {len(s2)}') #len() es una funcion que devuelve el largo de un string, se puede usar dentro de un f-string para imprimir el resultado

print (40*"-")
#Tipo de dato bool (booleano)

print("bool")
bo= True # un booleano puede ser True o False (con mayuscula)(significa verdadero o falso)
print(bool(3)) # cualquier numero diferente de 0 es considerado True, mientras que el numero 0 es considerado False
print(bo) 
```

## Operaciones Matemáticas Básicas

En python se pueden hacer sumas, restas, multiplicaciones, divisiones y más...  
Operaciones comunes:  
- Suma (+)  
- Resta (-)  
- Multiplicación (*)  
- División (/)  
- Potencia (**)  
- Módulo (%): Retorna el resto de la división  


*Ejemplo:*

```python
a = 10
b = 3

print(a + b)  # suma: 13
print(a - b)  # resta: 7
print(a * b)  # multiplicación: 30
print(a / b)  # división: 3.3333333333333335
print(a ** b) # potencia: 1000
print(a % b)  # módulo: 1, que es el resto de la división de 10 entre 3, ya que 10 dividido por 3 es 3 con un resto de 1
```

las opreraciones no se limitan solo a los números, sino a todos los tipos de datos

```python
nombre = "Juan"
apellido = "Pérez"
nombre_completo = nombre + " " + apellido # concatenación de strings con un espacio entre ellos
print(nombre_completo)
```
