# ¿Qué es un algoritmo?
Un algoritmo es una secuencia ordenada de instrucciones o pasos para resolver un problema. 
 En este caso sirven para que la computadora entienda lo que queremos hacer.  

*Analogía:* Piensa en una receta de cocina o en cómo preparas un té. Si cambias el orden (ej. echas el agua antes de calentarla), el resultado no es el mismo.


*Ejemplo: preparar té* 
- Calentar agua
- Poner el té en la taza
- Verter agua caliente
- Esperar 3 minutos
- Servir

```python
print("Hola mundo")
```

## *¿Qué hicimos?*

## Variables 

Las variables son espacios en la memoria que almacenan datos, al que se le asigna un valor o texto.

- **Modelo mental:** Imagina una caja con una etiqueta (el nombre de la variable). Lo que guardas dentro es el valor.
- Python asigna el tipo de dato automáticamente según lo que guardes. (no es necesitas declarar el tipo de variablen en otros lenguajes si es necesario)


**Tipos de datos básicos:**

| Tipo            | Ejemplo      | Que es?                                                                  |
| --------------  | ------------ | -------------------------------------------------------------------------|
| Entero (int)    | 5            | Son basicamente numeros (sin decimales)                                  |  
| Flotante (float)| 3.14         | Permite usar numeros decimales                                           |
| Cadena (str)    | "Hola"       | Nos permite crear cadenas de texto (Se escribe entre comillas.)          |
| Booleano (bool) | True / False | Es un tipo de dato que trabaja con dos estados, True y False (Verdadero/Falso).|


### Tipo `int` — Números enteros

```python
x = 1
y = 2
z = 200
print(x + y + z)  # 203
```

### Tipo `float` — Números con decimales

```python
f1 = 3.14
print(f1)           # 3.14

# Ojo: si sumas un int con un float, el resultado es float
resultado = x + f1
print(type(resultado))  # <class 'float'>
```

### Tipo `str` — Cadenas de texto

Un string es cualquier texto entre comillas (simples o dobles). 

Primero lo más básico: crear e imprimir un string.

```python
s1 = 'Hola mundo!'  # las comillas simples o dobles sirven para definir un string
print(s1)           
print(len(s1))      # 11  → len() es una funcion que devuelve el largo de un string


print(s1[5:]) #imprime desde el indice 5 hasta el final del string
print(s1[6:8]) #imprime desde el indice 6 hasta el indice 8 (sin incluir el indice 8)

s2 = s1 + '!!' # concatenacion de strings, se puede usar el operador + para unir dos strings
print(s2)
```

### Tipo `bool` — Booleanos

Solo puede ser `True` o `False`. Muy útil para tomar decisiones.

```python
bo = True
print(bo)         # True

# Cualquier número distinto de 0 se considera True; el 0 es False
print(bool(3))    # True
print(bool(0))    # False
```

## Operaciones Matemáticas Básicas

En python se pueden hacer sumas, restas, multiplicaciones, divisiones y más...  
Operaciones comunes:  
- Suma (+)  
- Resta (-)  
- Multiplicación (*)  
- División (/)  
- Potencia (**)  
- Módulo (%): Retorna el resto de la división  


*Ejemplo:*

```python
a = 10
b = 3

print(a + b)  # suma: 13
print(a - b)  # resta: 7
print(a * b)  # multiplicación: 30
print(a / b)  # división: 3.3333333333333335
print(a ** b) # potencia: 1000
print(a % b)  # módulo: 1, que es el resto de la división de 10 entre 3, ya que 10 dividido por 3 es 3 con un resto de 1
```

las opreraciones no se limitan solo a los números, sino a todos los tipos de datos

```python
nombre = "Juan"
apellido = "Pérez"
nombre_completo = nombre + " " + apellido # concatenación de strings con un espacio entre ellos
print(nombre_completo)
```
