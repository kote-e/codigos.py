# Ciclos y Estructuras Repetitivas

### ¿Qué es un Ciclo?
Un ciclo es una instrucción que permite repetir un bloque de código varias veces.
*Ejemplo*: Imagina que eres un atleta corriendo en una pista. Cada vuelta que das es una iteración. Sigues corriendo mientras el entrenador no toque el silbato (la condición).

Existen dos escenarios principales:
- Sabes cuántas vueltas darás: "Corre 10 vueltas".
- No sabes cuántas, pero tienes una condición: "Corre hasta que te canses".

estos son los 2 tipos de ciclos que existen:
- Los que se ejecutan una determinada cantidad de veces, conocida previamente.
- Los que se ejecutan mientras una condición se cumple.

## while

El ciclo while ejecuta un bloque de código mientras una condición sea verdadera (True).Las instrucciones que se repiten son las que se encuentran indentadas a la derecha, dentro de la instrucción
 Sintaxis:

```python
while condición:
    #instrucciones mientras la condición sea verdadera
```

**Tips:**  
- Iteración: Es el nombre técnico de cada "vuelta" del ciclo.
- Evaluación previa: La condición se revisa antes de entrar. Si la condición es falsa desde el inicio, el ciclo no se ejecuta ni una vez.
- El peligro del infinito: Si la condición nunca se vuelve falsa, el programa se quedará "pegado" para siempre. ¡Asegúrate de cambiar algo dentro del ciclo para que termine!.

````python
m = int(input("m: "))
n = int(input("n: "))
k = 0

#esto es un ciclo while que se ejecuta mientras m sea mayor a 0
while m > 0:
    m = m - 1 #se resta 1 a m en cada iteracion(contador)
    k = k + n #se suma n a k en cada iteracion (acumulador)

print(f"El producto de m y n es: {k}")
````

## Patrones Comunes: Las "Recetas" de Programación
Un patrón es una solución que sirve para muchos problemas diferentes. Aquí los más importantes:

| Patrón           | Objetivo                                  | Valor inicial | Operación clave                  |
|------------------|-------------------------------------------|---------------|----------------------------------|
| Suma             | Acumular un total                         | suma = 0      | suma = suma + n                  |
| Multiplicación   | Calcular factoriales o potencias          | prod = 1      | prod = prod * n                  |
| Conteo           | Contar cuántas veces pasa algo            | cont = 0      | cont = cont + 1                  |
| Máximo           | Encontrar el valor más grande             | mayor = -1    | if n > mayor: mayor = n          |
| Mínimo           | Encontrar el valor más pequeño            | menor = 100   | if n < menor: menor = n          |

*Nota sobre Máximo/Mínimo:* Usamos float("-inf") (infinito negativo) para el mayor y float("inf") (infinito positivo) para el menor. Esto asegura que el primer número que ingrese el usuario siempre sea el nuevo ganador.

````python
# este es un programa que calcula la suma de los numeros al cuadrado desde 1 hasta n
n = int(input("Ingrese n: "))
suma = 0
cont = 1
# este es un ciclo while que se ejecuta mientras cont sea menor o igual a n
while cont <= n:
    num = cont**2 # se calcula el cuadrado de cont y se guarda en num
    suma = suma + num
    cont += 1 # se incrementa cont en 1 en cada iteracion

print("La suma de los numeros al cuadrado es:", suma)
````

## Ciclos anidados

Puedes poner un ciclo dentro de otro. Cuando esto pasa, el ciclo interno debe terminar todas sus vueltas antes de que el ciclo externo pueda dar la siguiente.

*ejemplo:* Reloj: El ciclo externo son las horas y el interno los minutos. Los minutos deben dar 60 vueltas completas antes de que la hora avance solo una vez.

````python
#comprobar si un numero es primp con ciclo while
num = int(input("Ingrese un numero: "))
numero= 2
if num < 2:
    print(f"{num} no es un numero primo")
else:
    while numero < num:
        if num % numero == 0:
            print(f"{num} no es un numero primo")
            break
        numero += 1
    else:
        print(f"{num} es un numero primo")

print("-"*20)
# programa con ciclos anidados
n = int(input("Ingrese n: "))
while n > 0:
    m = int(input("Ingrese m: "))
    while m > 0:
        print("*", end="")
        m -= 1
    print()
    n -= 1  
````

*tips:* 
- La Bandera (flag): A veces la condición no es un número, sino un estado (como encendido = True). Usar banderas ayuda a controlar ciclos complejos.
- Ruteo: Si el código no funciona, dibujen una tabla con el valor de cada variable en cada vuelta. Es la única forma de "pensar" como la computadora.
- Indentación: Recuérdales que si el código no está movido a la derecha, Python creerá que está fuera del ciclo.

# for
A diferencia del while (que depende de una condición que puede ser eterna), el ciclo for está diseñado para recorrer una colección o un rango de elementos.

*ejemplo:* Imagina una cinta transportadora de maletas en un aeropuerto. El ciclo for toma una maleta, hace algo con ella, y pasa automáticamente a la siguiente hasta que no queda ninguna. No necesitas preguntarle al guardia si puedes seguir; la cinta simplemente se detiene cuando llega al final.

#### La función range(): El motor del for
solemos usar for junto con range() para generar una secuencia de números. Sintaxis: ``` range(inicio, fin, salto) ```

- Inicio: ¿Desde dónde empezamos? (Por defecto es 0).
- Fin: ¿Hasta dónde llegamos? ¡Ojo! El ciclo se detiene justo antes de este número.
- Salto: ¿De cuánto en cuánto avanzamos? (Por defecto es 1).

| Código             | Resultado           | Explicación                                           |
|--------------------|---------------------|-------------------------------------------------------|
| range(5)           | 0, 1, 2, 3, 4       | Empieza en 0 y llega hasta el 4 (5 números en total). |
| range(2, 6)        | 2, 3, 4, 5          | Empieza en 2 y termina justo antes del 6.             |
| range(1, 10, 2)    | 1, 3, 5, 7, 9       | Empieza en 1 y va de 2 en 2.                          |

#### Recorriendo Strings
El for es muy potente porque puede tratar a una palabra como una lista de letras.


````python
nombre = "Python"
for letra in nombre:
    print("Dame una", letra)
````

*Nota:* La variable letra es una variable ```temporal```. En cada vuelta, toma el valor del siguiente elemento de la secuencia automáticamente.

#### Patrones con for


````python
#El Acumulador (Suma de los primeros n números)

n = int(input("¿Hasta qué número sumar? "))
suma = 0
for i in range(1, n + 1): # Usamos n + 1 para incluir al número n
    suma += i
print("La suma total es:", suma)


#El Contador  (Contar cuántas letras 'a' tiene una frase)
frase = input("Escribe algo: ")
contador = 0
for caracter in frase:
    if caracter == "a":
        contador += 1
print("Encontré", contador, "letras 'a'")
````

## ¿Cuándo usar for y cuándo while?

| Característica | Ciclo for                              | Ciclo while                                      |
|----------------|----------------------------------------|--------------------------------------------------|
| Control        | Basado en una colección o rango.       | Basado en una condición lógica.                  |
| Uso ideal      | Cuando sabes el número de repeticiones.| Cuando no sabes cuándo terminará.                |
| Riesgo         | Es muy difícil que sea infinito.       | Si olvidas actualizar la variable, es infinito.  |
| Ejemplo        | "Limpia estos 10 platos".              | "Limpia platos hasta que no haya más".           |
